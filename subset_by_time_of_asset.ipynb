```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from IPython.display import Markdown,display
import datetime
import time

file_path = 'data/data_by_asset_keys/'
file_name = '4.csv'
df = pd.read_csv(file_path+file_name)
```

```python
subset_keys = ['recorded_at','MDI_OBD_FUEL','ODO_FULL_METER']
subset = df[subset_keys]
subset
```

```python
df_list = []
count_data = len(subset)


start_line = 0
last_time = time.mktime( datetime.datetime.strptime(subset.iloc[0]['recorded_at'], "%Y-%m-%dT%H:%M:%SZ").timetuple() )
for i in range(1,count_data):
    last_time = time.mktime( datetime.datetime.strptime(subset.iloc[i-1]['recorded_at'], "%Y-%m-%dT%H:%M:%SZ").timetuple() )
    current_time   = time.mktime( datetime.datetime.strptime(subset.iloc[i]['recorded_at'], "%Y-%m-%dT%H:%M:%SZ").timetuple() )
    if current_time-last_time > 1800:
        df_list.append(subset.iloc[start_line:i])
        start_line = i     
    if i == count_data-1:    
        df_list.append(subset.iloc[start_line:count_data])
```
